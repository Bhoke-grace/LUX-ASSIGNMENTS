```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from datetime import datetime
```

```python
#ALT Z-FIT IN THE SCREEN
df=pd.read_csv(r'C:\Users\User\Desktop\DATA SCIENCE FINALYYYY\PYTHON\train_data_cross-sell.csv', encoding= "latin1")
df.head()
```

```python
#Reducing the no. of columns to reduce higgh dimensionality
df=df.drop(['ÿid', 'Customer_Residence_Sub_County', 'Agent_name'], axis=1)
df.head()
```

```python
df['Customer_Date_of_Birth']=pd.to_datetime(df['Customer_Date_of_Birth'])
df['Life_policy_start_date']=pd.to_datetime(df['Life_policy_start_date'])
df.head()
```

```python
#usefrom datetime import datetime

today=datetime.today()
print("Today's date is:", today)
```

```python
df['Age']=today.year -df['Customer_Date_of_Birth'].dt.year
df.head()
```

```python
#calculating the policy tenure in years
df['Policy Tenure']=today.year -df['Life_policy_start_date'].dt.year
df.head()
```

```python
# Calculate the policy tenure in months
df['Policy_Tenure'] = (today.year - df['Life_policy_start_date'].dt.year) * 12 + (today.month - df['Life_policy_start_date'].dt.month)

# Display the first few rows of the DataFrame
df.head()
```

```python
#Droping columns
df=df.drop(['Customer_Date_of_Birth', 'Life_policy_start_date', 'Policy Tenure' ], axis=1)
df.head()
```

```python
#import the library that performs label encodeing
from sklearn.preprocessing import LabelEncoder

#get the no. of unique counts for each column in our data frame
unique_counts=df.nunique()

#select only colums w 2 or less unique counts
col_to_encode=unique_counts[unique_counts<=2].index.to_list()

#Initiialize the label encoder
le=LabelEncoder()

#Transform the selected colums
for col in col_to_encode:
    df[col]=le.fit_transform(df[col])

df.head()
```

```python
#separate the dependant variable and the indep. variables
y=df['Response']

x=df.drop('Response', axis=1)
x.head()
```

```python
#correlation of x & y
target_correlation=x.corrwith(y)
print(target_correlation)
#multiply by 100
```

```python
#correlations btwn xs
x_correlations=x.corr()
print(x_correlations)
```

```python
#plotting heart
plt.figure(figsize=(9,6))
sns.heatmap(x_correlations, annot=True, cmap='viridis')
#color viridis, blues, coolwarms
plt.title('CORRELATION MATRIX')
plt.show()
     
```
